## Matplotlib
- Matplotlib is an amazing visualization library in Python for 2D plots of arrays. 
- Matplotlib is a multi-platform data visualization library built on NumPy arrays.
- Matplotlib is open source and we can use it freely.

### Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(11,16)
y=x*2
print("x values=",x)
print("y values=",y)
```

#### Line plot:

```python
#line plot
plt.plot(x,y)
```

#### Scatter plot:

```python
plt.scatter(x,y)
```

#### Bar plot:

```python
plt.bar(x,y)
```

#### Barh plot:

```python
plt.barh(x,y)
```

#### Pie plot:

```python
plt.pie(x)
```

```python
plt.pie(y)
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```

## Matplotlib label::

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Creating labels')
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Creating labels')
plt.suptitle('Matplotlib label')
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
font1 = {'family':'serif','color':'blue','size':20}          ##changing fonts
plt.title('Creating labels',fontdict = font1)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
font1 = {'family':'serif','color':'blue','size':20}          ##changing fonts
plt.title('Creating labels',fontdict = font1,loc = 'left')
```

## Matplotlib Grid:

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Creating labels',fontdict = font1,loc = 'left')
plt.grid()
plt.show()                 ##grid lines
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Creating labels',fontdict = font1,loc = 'left')
plt.grid(axis = 'x')
plt.show()     
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Creating labels',fontdict = font1,loc = 'left')
plt.grid(axis = 'y')
plt.show()     
```

## Matplotlib subplots:
- Multiple plots in one figure

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```
