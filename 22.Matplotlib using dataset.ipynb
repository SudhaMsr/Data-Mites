## Object oriented plotting

```python
fig=plt.figure()
ax=fig.add_axes([0.5,0.2,0.8,0.9])
ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.5,0.2,0.8,0.9])
ax.plot(x,y)
ax.set_xticks([10,15])
plt.show()

## ticks: array-like containing the list of xtick locations. Passing an empty list removes all xticks. 
```

```python
fig=plt.figure()   ##step1
ax=fig.add_axes([0.5,0.2,0.8,0.9])  ##left,bottom,width,height #step 2
ax.plot(x,y)  #step 3
ax.set_xlim([12,14])
ax.set_xticks([12,14])
ax.set_ylim([25,29])
ax.set_yticks([24,30])
plt.show()
```

# Generating Multiple plots

```python
fig=plt.figure(figsize=(10,15))
fig,ax=plt.subplots(2,3)
ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x,y)
ax[1,1].pie(x)
ax[1,2].barh(x,y)
#ax[2,0].plot(x,y)
#ax[2,1].pie(x,y)
#ax[2,2].barh(x,y)
plt.tight_layout(pad=1)
#plt.savefig('multiple plots.png',format='png')
plt.show()
```

```python
fig=plt.figure(figsize=(5,10))
fig,ax=plt.subplots(2,3)
ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x,y)
ax[1,1].pie(x)
ax[1,2].barh(x,y)
ax[0,0].plot(x,y)
ax[1,1].pie(x,y)
ax[1,2].barh(x,y)
plt.tight_layout(pad=1)
#plt.savefig('multiple plots.png',format='png')
plt.show()
```

# Plot within plot

```python
x=np.arange(1,11)
y=x**2
print(x,y)
fig2=plt.figure()
#creating main axes- values blw 0 & 1
axes_main=fig2.add_axes([0.1,0.1,0.9,0.9])  #L,B,W,H

#creating inset axes- values b/w 0 & 1
axes_inset=fig2.add_axes([0.3,0.2,0.4,0.5])

#change the numbers in axes inset and see the difference
#main figure
axes_main.plot(x,y,'b')
axes_main.set_xlabel('X_label_main')
axes_main.set_ylabel('Y_label_main')
axes_main.set_title('Title_main')

#inset figure
axes_inset.plot(x,y,'r')
axes_inset.set_xlabel('X_label_main')
axes_inset.set_ylabel('Y_label_main')
axes_inset.set_title('Title_main')

plt.show()
```
