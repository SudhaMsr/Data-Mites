# Strings:
* Strings in python are surrounded by either single quotation marks, or double quotation marks.
   * 'hello' is the same as "hello".
* We can print a string literal with the print() function:
* String is a data type in python
* String is a sequence of characters enclosed in quotes

```python
print("Hello, Alaric!")
print('Hello, Alaric!')
```

## Assigning a string to variable

```python
a = "Hello, Alaric!"
print(a)
```

## Multiline strings 

```python
a = """Hi Everyone!
I'm Alaric from India
bye..:)"""
print(a)
```

## String declaration

```python
str="Hello, Alaric!"
print(str)
print(type(str))        #Type of data
print(len(str))         #Length of string (spaces are also included)
```

## String fuctions:
### 1. len() -Used to returns the length of a string

```python
a= ("alaric")
len(a)
```

### 2. string.endswith(ric) -This function tells you whether the function is ending with "ric" or not. Since alaric ends wirh ric

```python
a.endswith("ric")
```

### 3. string.count()- counts the total number of characters in string.

```python
a.count(a)
```

### 4. string.capitalize()- It capitalizes the first letter of string. 

```python
a.capitalize()
```

### 5. string.find(word)- This fuction finds the word & returns the index of word 

```python
a.find(a)
```

### 6. string.relace(old word,new word)
